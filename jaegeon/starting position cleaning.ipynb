```python
import pandas as pd
import numpy as np

```

```python
start_pos = pd.read_csv('C:\\Users\luker\OneDrive\Desktop\erdos_ds_f1\data_f1db\\f1db-races-starting-grid-positions.csv')
race_pos = pd.read_csv('C:\\Users\luker\OneDrive\Desktop\erdos_ds_f1\cleaned_data\\f1db-races-race-results-NO-DNF.csv')

race_pos_df = race_pos[['year', 'raceId', 'driverId', 'positionNumber']]
```

```python
# Merge start_pos and race_pos on both 'raceId' and 'driverId'
merged_data = pd.merge(start_pos, race_pos_df, on=['raceId', 'driverId'], suffixes=('_start','_race'))
```

```python
# Compute the absolute difference between starting position and race position

merged_data['position_diff'] = np.abs(merged_data['positionNumber_start'] - merged_data['positionNumber_race'])
```

```python
merged_data.head()
```

```python
# Rename colums for clarity
merged_data.rename(columns={
    'positionNumber_start': 'starting_position',
    'positionNumber_race' : 'race_position',
    'position_diff' : 'absolute_position_diff'
}, inplace=True)
```

```python
# Only keep relevant columns
merged_data.rename(columns={
    'year_start' : 'year'
}, inplace=True)

start_race_pos_diff_noDNF = merged_data[['year', 'raceId', 'driverId', 'starting_position', 'race_position', 'absolute_position_diff']]
```

```python
# Export start_race_pos_diff_noDNF to CSV

start_race_pos_diff_noDNF.to_csv('start-to-race-position-difference-No-DNF.csv')
```

```python
# Dataframe with mean absolute position difference per race

mean_pos_diff = start_race_pos_diff_noDNF.groupby(['raceId']).mean(numeric_only=True)
```
